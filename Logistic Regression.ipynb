### Alanine aminotransferase dropped as an aminotransferase (aspartate aminotransferase) is already present

```python
liver_disease_transformed = liver_disease_transformed.drop(columns=['Alanine_Aminotransferase'])
```

# Scaling features to aid in grid search solver convergence

```python
features_unchanged = ['Gender', 'Outcome']
features_to_scale = [col for col in liver_disease_transformed.columns if col not in features_unchanged] 

scaled_features = liver_disease_transformed[features_to_scale]
original_features = liver_disease_transformed[features_unchanged]

#Importing the scaling function 
from sklearn.preprocessing import StandardScaler 

#Initialise scaling function 
scaler = StandardScaler()


scaled_data = scaler.fit_transform(scaled_features)

#Converting output (Series) to Data Frame 

scaled_data = pd.DataFrame(scaled_data, columns=features_to_scale)

liver_disease_scaled = pd.concat([original_features, scaled_data], axis=1)

liver_disease_scaled.head()
```

# MODEL BUILDING 

```python
X = liver_disease_scaled.drop(['Outcome'], axis='columns')
y = liver_disease_scaled['Outcome']
```

```python
# Split data into training and testing sets 

from sklearn.model_selection import train_test_split, GridSearchCV 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
# Import the model 

from sklearn.linear_model import LogisticRegression 


# Instantiate the model       

logreg = LogisticRegression(random_state=7, max_iter=500)
```

### Tweak: max_iter in logreg increased to 500 from default of 100 when grid search solver did not converge 

```python
#Grid search cross validation

parameters = [{'solver': ['lbfgs', 'liblinear', 'sag', 'saga', 'newton-cg']}] 


grid_search = GridSearchCV(estimator = logreg,
                           param_grid = parameters,     #list of evalution metrics 
                           refit=True
                           )


grid_search.fit(X_train,y_train)

print(f"'Best Solver:{grid_search.best_params_}")
```

### Setting solver and tuning remaining hyperparameters

```python
logreg = LogisticRegression(solver='lbfgs', random_state=7, max_iter=500)
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, accuracy_score, make_scorer
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc, roc_auc_score
```

```python
parameters = [{'penalty':['none', 'l2'],
              'C':[0.001, 0.01, 0.1, 1, 10, 100, 1000]}]

scoring = {"AUC": "roc_auc", "accuracy": 'accuracy'}

grid_search = GridSearchCV(estimator = logreg,
                           scoring = scoring,
                           return_train_score = True,
                           param_grid = parameters,
                           cv = 10,
                           n_jobs = 2,
                           refit = 'AUC')

grid_search.fit(X, y)
results = grid_search.cv_results_

print('='*50)
print(f"Best parameters: {grid_search.best_params_}\nBest score: {grid_search.best_score_}")
print('='*50)
```

```python
logreg = LogisticRegression(solver='lbfgs', penalty='l2', C=10, random_state=7, n_jobs=None, l1_ratio=None)
```

```python
from sklearn.model_selection import KFold, cross_validate, cross_val_score

scoring = {"AUC": "roc_auc", 'accuracy': 'accuracy'}

results = cross_validate(logreg, X, y, cv=10, return_train_score=False, scoring=list(scoring.values()))


print('10-fold cross-validation results:')
for sc in range(len(scoring)):
    print(logreg.__class__.__name__+" average %s: %.3f (+/-%.3f)" % (list(scoring.keys())[sc], -results['test_%s' % list(scoring.values())[sc]].mean(), results['test_%s' % list(scoring.values())[sc]].std()))
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, shuffle=True, random_state=42)

training_scores =[]
testing_scores = []

# Iterate over each fold

for train_index, test_index in kfold.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit the model on the training data
    model = logreg.fit(X_train, y_train)

    # Evaluate the model and store the accuracy score
    training = logreg.score(X_train, y_train)
    testing = logreg.score(X_test, y_test)
    training_scores.append(training)
    testing_scores.append(testing) 

for i, (train, test) in enumerate(zip(training_scores, testing_scores), 1):
    print(f'Fold {i}: Training set accuracy = {train:.4f}, Testing set accuracy = {test:.4f}')

print(f'Average Training set accuracy: {np.mean(training_scores):.4f}')
print(f'Average Testing set accuracy: {np.mean(testing_scores):.4f}')
```

## Average scores of training and test sets are comparable, ruling out overfitting

```python
# Selecting Fold 1 because of highest testing set accuracy 

fold_number = 1

for i, (train_index, test_index) in enumerate(kfold.split(X, y), 1):
    if i == fold_number:
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

logreg.fit(X_train, y_train)
```

# Model Evaluation 

## Confusion matrix 

```python
from sklearn.metrics import confusion_matrix

y_pred_test = logreg.predict(X_test)

cm = confusion_matrix(y_test, y_pred_test)

print('Confusion matrix\n\n', cm)

print('\nTrue Negatives(TN) = ', cm[0,0])

print('\nTrue Positives(TP) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11, 8))

confusion_matrix = pd.DataFrame(data=cm, columns=['Negative', 'Positive'],
                                index=['Negative', 'Positive'])

sns.heatmap(confusion_matrix, annot=True, fmt='', cmap='Blues', square=True)

ax.xaxis.tick_top()
ax.set_title('Predicted', pad=15)
plt.ylabel('Actual')
plt.show()
```

## Classification Report 

```python
from sklearn.metrics import classification_report

target_names = ['Liver disease absent', 'Liver disease present']

print(classification_report(y_test, y_pred_test, target_names=target_names))
```

## Receiver Operating Curve 

```python
# Calculate AUC 

AUC = roc_auc_score(y_test, y_pred_test)

#Plot ROC

fpr, tpr, thresholds = roc_curve(y_test, y_pred_test, pos_label=True)

plt.figure(figsize=(10,10))

plt.plot(fpr, tpr, linewidth=2, label="AUC : {:.4f}".format(AUC))

plt.plot([0,1], [0,1], 'k--' )

plt.rcParams['font.size'] = 25

plt.title('Receiver Operating Curve for Predicting Liver Disease', fontsize=30, pad=30)

plt.xlabel('False Positive Rate' , fontsize=20)

plt.ylabel('True Positive Rate', fontsize=20)

plt.legend(loc=4)

plt.show()
```
